```python
import pandas as pd
import seaborn as sns
```

```python
pg = sns.load_dataset("penguins")
pg.head()
```

```python
g=sns.PairGrid(pg)
g.map(sns.scatterplot)
```

```python
sns.scatterplot(data=pg, x="flipper_length_mm", y = "bill_depth_mm")
```

```python
sns.scatterplot(data=pg, x="flipper_length_mm", y = "bill_depth_mm", hue="island")
```

```python
sns.scatterplot(data=pg, x="flipper_length_mm", y = "bill_depth_mm", hue="sex")
```

```python
sns.scatterplot(data=pg, x="flipper_length_mm", y = "bill_depth_mm", hue="species")
```

```python
sns.scatterplot(data=pg, x="species", y = "island")
```
